## BASIC TRANSFORMATIONS

Simple transformation methods

```python
#Global variables

black_cat = "https://raw.githubusercontent.com/lcarli/computer_vision/master/Images/Cat.jpg"
color_cat = "https://raw.githubusercontent.com/lcarli/computer_vision/master/Images/Color%20Cat.jpg"
```

```python
#BASIC IMPORTS

from pylab import *
from PIL import Image
import requests
from numpy import *
```

```python
#Opening image

image = Image.open(requests.get(black_cat, stream=True).raw)

display(image)
```

```python
#Thumbnails
im_thumb = image.thumbnail((128,128))

display(im_thumb)
```

```python
#Crop region
box = (100,100,150,150) #(left, upper, right, lower)
im_region = image.crop(box)

display(im_region)
```

```python
#Resize
im_resized = image.resize((128,128))

display(im_resized)
```

```python
#Rotate
im_rotated = image.rotate(45)

display(im_rotated)
```
