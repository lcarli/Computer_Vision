```python
#Global variables

black_cat = "https://raw.githubusercontent.com/lcarli/global-ai-2020-CV/master/Images/Cat.jpg"
color_cat = "https://raw.githubusercontent.com/lcarli/global-ai-2020-CV/master/Images/Color%20Cat.jpg"
```

```python
#BASIC IMPORTS

from pylab import *
from PIL import Image
import requests
from numpy import *
```

```python
#Opening image

image = Image.open(requests.get(black_cat, stream=True).raw)

display(image)
```

```python
# The file format of the source file.
print(image.format)

# The pixel format used by the image. Typical values are "1", "L", "RGB", or "CMYK."
print(image.mode)

# Image size, in pixels. The size is given as a 2-tuple (width, height).
print(image.size)

# Colour palette table, if any.
print(image.palette)
```

```python
#Working as array

im = array(image)
print(im.shape)
print(im.dtype)

imshow(im)
```

## BASIC TRANSFORMATIONS

Simple transformation methods

```python
#Thumbnails
im_thumb = image.thumbnail((128,128))

display(im_thumb)
```

```python
#Crop region
box = (100,100,150,150) #(left, upper, right, lower)
im_region = image.crop(box)

display(im_region)
```

```python
#Resize
im_resized = image.resize((128,128))

display(im_resized)
```

```python
#Rotate
im_rotated = image.rotate(45)

display(im_rotated)
```
